```python
import cv2
import pywt
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage,signal
from scipy import stats as stats
from skimage import feature
from skimage.feature import greycoprops,greycomatrix
from sklearn.preprocessing import MinMaxScaler
import os
import skimage.measure   
```

```python
images=[]
folder='./dataset/severe'
for filename in os.listdir(folder):
    img = cv2.imread(os.path.join(folder,filename))
    if img is not None:
        img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        img=cv2.resize(img,(28,28))
#         ret2,img = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        images.append(img)
energy_y=[]
contrast_y=[]
dissimilarity_y=[]
homogeneity_y=[]
correlation_y=[]
asm_y=[]
entropy_y=[]
```

```python
from statistics import median
g_energy=[]
g_contrast=[]
g_dissimilarity=[]
g_homogeneity=[]
g_correlation=[]
g_asm=[]
g_entropy=[]
for img in images:
    coeffs=pywt.wavedec(img, 'db1',level=1)
    cA=coeffs[0]
    cA = cA / cA.max()
    cA = 255 * cA 
    cA=cA.astype(np.uint8)
    g=greycomatrix(cA,[1],[0,np.pi/4,np.pi/2,3*np.pi/4,np.pi],normed=True)
    g_energy.append(np.mean(greycoprops(g, 'energy')))
    g_entropy.append((skimage.measure.shannon_entropy(g)))
    g_contrast.append(np.mean(greycoprops(g, 'contrast')))
    g_dissimilarity.append(np.mean(greycoprops(g, 'dissimilarity')))
    g_homogeneity.append(np.mean(greycoprops(g, 'homogeneity')))
    g_correlation.append(np.mean(greycoprops(g, 'correlation')))
    g_asm.append(np.mean(greycoprops(g, 'ASM')))
print("Using GLCM: \n")
print("Energy: ",median(g_energy)*100)
print("Entropy: ",median(g_entropy)*10000)
print("Contrast: ",median(g_contrast))
print("Dissimilarity: ",median(g_dissimilarity))
print("Homogeneity: ",median(g_homogeneity)*100)
print("Correlation: ",median(g_correlation)*100)
print("ASM: ",median(g_asm)*100)
energy_y.append(median(g_energy)*100)
contrast_y.append(median(g_contrast))
dissimilarity_y.append(median(g_dissimilarity))
homogeneity_y.append(median(g_homogeneity)*100)
correlation_y.append(median(g_correlation)*100)
asm_y.append(median(g_asm)*100)
entropy_y.append(median(g_entropy)*10000)
```
